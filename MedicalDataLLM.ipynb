```python
import csv
import requests
from bs4 import BeautifulSoup

csv_filename_input = 'netmeds_links.csv'

with open(csv_filename_input, 'r', encoding='utf-8') as file:
    csv_reader = csv.reader(file)
    header = next(csv_reader)  # Skip the header
    urls = [row[0] for row in csv_reader]

titles_to_scrape = [
    'INTRODUCTION',
    'USES OF THE TABLET',
    'HOW ATREST THE TABLET WORKS',
    'DIRECTIONS FOR USE',
    'SIDE EFFECTS OF THE TABLET',
    'COMMON',
    'UNCOMMON',
    'RARE',
    'HOW TO MANAGE SIDE EFFECTS',
    'WARNING & PRECAUTIONS'
]

for url in urls:
    try:
        response = requests.get(url)
        if response.status_code == 200:
            html = response.text
            soup = BeautifulSoup(html, 'html.parser')

            tablet_name = soup.find("h1", {"class": "black-txt"})
            if tablet_name:
                tablet_name = tablet_name.text.strip()
            else:
                print(f"No tablet name found for {url}. Skipping...")
                continue

            drug_info = soup.find_all("div", {"class": "druginfo_cont"})

            csv_filename_output = 'scrappingData.csv'
            with open(csv_filename_output, 'a', newline='', encoding='utf-8') as csv_file:
                csv_writer = csv.writer(csv_file)

                if csv_file.tell() == 0:
                    header = ['tablet_name'] + titles_to_scrape
                    csv_writer.writerow(header)

                data_dict = {'tablet_name': tablet_name}

                for section in drug_info:
                    title = section.find("h2")
                    if title:
                        title = title.text.strip()
                    else:
                        continue

                    content = section.find("div", {"class": "inner-content"})
                    if content:
                        content = content.text.strip()
                    else:
                        continue

                    if title in titles_to_scrape:
                        data_dict[title] = content

                csv_writer.writerow([data_dict.get(col, '') for col in header])

            print(f"Data saved to {csv_filename_output}")

        else:
            print(f"Failed to fetch data from {url}. Status code: {response.status_code}")

    except requests.RequestException as e:
        print(f"Failed to fetch data from {url}. Error: {e}")
```
